# urllib解析

````python
import urllib.request
#获取一个get请求
response = urllib.request.urlopen("http://www.baidu.com")
#将获取到的网页信息解码
#response.read().decode("utf-8")
````

````python
#获取一个post请求
import urllib.parse
data = bytes(urllib.parse.urlencode({"hello":"world"}),encoding="utf-8")
response_1 = urllib.request.urlopen("http://httpbin.org/post",data=data)
response_1.read().decode("utf-8")
````

````python
#超时处理
try:
    response_2 = urllib.request.urlopen("http://httpbin.org/get",timeout=0.01)
    response_2.read().decode("utf-8")
except Exception as e:
    print("time out")
````

````python
#网页内容的深度解析
response_3 = urllib.request.urlopen("http://httpbin.org/get")
#response_3.status # 200
response_3.getheaders()
````

````python
#伪装成浏览器  不让被发现418
url = "http://httpbin.org/post"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36 Edg/91.0.864.70"
}
data = bytes(urllib.parse.urlencode({"name":"eric"}),encoding="utf-8")
req = urllib.request.Request(url=url,data=data,headers=headers,method="POST")
response_4 = urllib.request.urlopen(req)
response_4.read().decode("utf-8")
````

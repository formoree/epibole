```python
#读取数据
import pandas as pd
data = pd.read_table("DEG10n.txt",header=None)
```

```python
#将数据转化为list
data1 = data.values.tolist()
```

```python
data1
```

```python
#定义函数提取list中字符串
def translate_list(l):
    s = ''
    for i in range(len(l)):
        s+=l[i]
    return s 

#提取关键value
idx =[]
se = []
ses = ''
for i in data1:
    if "E" in str(i):
        i = str(i)[3:-2]
        idx.append(i)
        se.append(ses)
        ses = ''
    else:
        ses+=translate_list(i)

se = se[1:]

#打包为字典
data = zip(idx,se)
idx
```
